```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from sklearn import datasets
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
n_pts = 500
X, y = datasets.make_circles(n_samples=n_pts,random_state=100,noise=0.1,factor=0.2)
```

```python
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1])
```

```python
model = Sequential()
model.add(Dense(4,input_shape=(2,),activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
model.compile(Adam(lr=0.01),'binary_crossentropy',metrics=['accuracy'])
```

```python
h = model.fit(x=X,y=y,verbose=1,batch_size=20,epochs=100,shuffle='true')
```

```python
plt.plot(h.history['accuracy'])
plt.xlabel('epochs')
plt.legend(['accuracy'])
plt.title('accuracy')
```

```python
plt.plot(h.history['loss'])
plt.xlabel('epochs')
plt.legend(['loss'])
plt.title('loss')
```

```python
 def plt_decision_boundry(X,y,model):
    x_span = np.linspace(min(X[:,0])-1,max(X[:,0])+1)
    y_span = np.linspace(min(X[:,1])-1,max(X[:,1])+1)
    xx, yy = np.meshgrid(x_span,y_span)
    xx_,yy_ = xx.ravel(),yy.ravel()
    grid = np.c_[xx_,yy_]
    prediction = model.predict(grid)
    z = prediction.reshape(xx.shape)
    plt.contourf(xx,yy,z)
```

```python
plt_decision_boundry(X,y,model)
plt.scatter(X[:n_pts,0],X[:n_pts,1])
plt.scatter(X[n_pts:,0],X[n_pts:,1])
```

```python
plt_decision_boundry(X,y,model)
plt.scatter(X[:n_pts,0],X[:n_pts,1])
plt.scatter(X[n_pts:,0],X[n_pts:,1])
x = 0.1
y = 0.75
point = np.array([[x,y]])
prediction = model.predict(point)
plt.plot([x],[y],marker="o",markersize=10,color="red")
print("prediction is:", prediction)
```
